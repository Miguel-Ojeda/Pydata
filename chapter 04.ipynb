## Simulación de caminos aleatorios con numpy
Let’s first consider a simple random walk starting at 0 with steps of 1 and –1 occurring with equal probability.

Here is a pure Python way to implement a single random walk with 1,000 steps using the built-in random module:

```python
#! blockstart
import random
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
#! blockend
```

```python
plt.plot(walk[:100])
```

```python
# COn numpy
# Generamos, en un solo paso, los 1000 números aleatorios
nsteps = 1000
rng = np.random.default_rng(seed=12345)  # fresh random generator

draws = rng.integers(0, 2, size=nsteps)  # o sea, al azar se extrae un 0 o un 1 !!  (el 2 no se incluye!!)
# traducimos ahora... si sale 0, pues colocamos un 1, si no, pues un -1
steps = np.where(draws == 0, 1, -1)

# calculamos el camino, simplemente acumular
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

Imaginamos ahora que queremos calcular el primer 'crossing time', o sea, la primera vez que el camino llega a un valor determinado.

Por ejemplo, cuánto tiempo le llevó, en qué paso ocuurrió, en que alcanzamos al distancia 10 desde el origen...

np.abs(walk) >= 10 gives us a boolean array indicating where the walk has reached or exceeded 10, pero nosotros lo que queremos es el índice del primer valor que alcanza
el 10 o el -10.

Podemos para ello calcular el vector Booleano que vale True cuando el valor es 10 o -10, y a ese vector aplicar **argmax** que nos de el primer índice en el que ocurre el máximo (en nuestro caso, el primer índice en que hay True!!)

```python
bool_array = (np.abs(walk) == 10)

primer_indice = bool_array.argmax()
primer_indice
```

### Como podemos hacer muchas simulaciones de estos caminos a la vez con numpy, por ejemplo, 5000 caminos!!!

```python
nwalks = 5_000
nsteps = 1_000
draws = rng.integers(0, 2, size=(nwalks, nsteps))

steps = np.where(draws > 0, 1, -1)

walks = steps.cumsum(axis=1)
walks
```

```python
walks[0]
```

```python
walks.max()
```

```python
walks.min()
```

Out of these walks, let’s compute the minimum crossing time to 30 or –30. This is slightly tricky because not all 5,000 of them reach 30. We can check this using the any method:

```python
hits30 = (np.abs(walks) == 30).any(axis=1)
```

```python
hits30.shape
```

```python
hits30
# Nos dice si cada uno de los 5000 caminos llega a tener el valor 30!!
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis=1)
```
